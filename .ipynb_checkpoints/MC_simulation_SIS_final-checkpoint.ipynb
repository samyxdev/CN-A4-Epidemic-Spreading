# Monte Carlo simulation of SIS epidemic spreading


### Import libraries

```python
import igraph as ig
import numpy as np
import time
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('ggplot')

%load_ext cython
```

### Simulation function
The function below performs 1 full simulation of SIS epidemic spreading.

```python
%%cython
import numpy as np
import igraph as ig
from libc.math cimport pow
from libc.stdlib cimport rand, srand, RAND_MAX
from libc.time cimport time,time_t

def simulation(g, int len_g, char[:] state, double mu, double beta):
    
    """
    Applies one simulation to the graph in argument and returns 
    the average infection rate over the last 100 time steps
    """ 
    
    # Constants
    cdef int TMAX = 1000
    cdef int TTRANS = 900
    cdef double _beta = 1 - beta
    
    # Declared variables
    cdef double[:] infected_ratio_list = np.zeros(TMAX - TTRANS)
    cdef long start_time
    cdef int i, j, v, n, infected_vertices
    cdef char[:] old_state = np.zeros(len_g, dtype = np.int8)
    cdef double prob
    cdef double a = RAND_MAX
    cdef int num_infected = 0
    
    # set random seed 
    srand(time(NULL)) 
    
    #simulation loop
    for i in range(TMAX):
        
        if i >= TTRANS:
            num_infected = 0
            
        # copy old state
        for j in range(len_g):
            old_state[j] = state[j]
        #old_state = np.copy(state)
        
        # update states of all nodes
        for j in range(len_g):
            
            # if old node state is I
            if old_state[j]:
                prob = rand() / a # generate random number in [0, 1]
                if prob < mu: # update new state
                    state[j] = 0 
                else:
                    state[j] = 1
                
            # if old node state is S
            else:
                infected_vertices = 0  # counter of the infected vertices of current SUCSECTIBLE node
                
                for n in g.neighbors(j):
                    infected_vertices += old_state[n]
                #infected_vertices = sum([old_state[v] for v in g.neighbors(j)]) # number of infected neighbors
                prob = rand() / a
                if prob < pow(1 - beta, infected_vertices):
                    state[j] = 0
                else:
                    state[j] = 1
                    
            # we compute infection ratio from iteration 900
            if i >= TTRANS:
                num_infected += state[j]
            
        if i >= TTRANS:
            infected_ratio_list[i - TTRANS] = num_infected/len_g

        
    return np.mean(np.array(infected_ratio_list))
```

## Example with a simple network

### Create network

```python
len_g = 500
g = ig.Graph.Barabasi(len_g, 5)
```

### Parameters

```python
MU = 0.5
BETA = 0.6
P0 = 0.2
```

### Perform simulation

```python
state = np.array(np.random.binomial(1, P0, size=len_g), dtype=np.int8) # initial state

avg_infection_ratio = simulation(g, len_g, state, MU, BETA)
```

```python
avg_infection_ratio
```

### Perform n simulations
Let's define an auxiliary function for performing NREP simulations


```python
NREP = 50 
```

```python
def perform_n_iterations(g, len_g, MU, BETA, n=NREP):
    start_t = time.time()
    
    infected_ratio_list = np.zeros(NREP) # vector with avg infection ratio of each simulation
    for i in range(NREP):
        start_time = time.time()
        state = np.array(np.random.binomial(1, P0, size=len_g), dtype=np.int8) # initial state
        infected_ratio_list[i] = simulation(g, len_g, state, MU, BETA)
        print("Time required by simulation {}: {} seconds".format(i+1, time.time()-start_time))
    print("\nTime required by 100 simulations: {} seconds\n\n".format(start_t - time.time()))
    
    return np.mean(infected_ratio_list) # return average of averages over 100 iterations
```

```python
a = perform_n_iterations(g, len_g, MU, BETA, NREP)
```

```python
a
```

```python
# We need now to iterate over different beta values
#beta_list = np.linspace(0.01, 1, int((1-0.01)/STEP_BETA))
beta_list = [0.1, 0.2, 0.4, 0.6, 0.8, 0.9]
```

```python
def iterate_over_beta(g, len_g, MU, beta_list):
    avg_over_beta = np.zeros(len(beta_list))
    for i, beta in enumerate(beta_list):
        print("Calculate value for beta = {}".format(beta))
        avg_over_beta[i] = perform_n_iterations(g, len_g, MU, beta, NREP)
    
    return avg_over_beta # returns the value for each beta value given the MU value
```

```python
avg_over_beta = iterate_over_beta(g, len_g, MU, beta_list)
```

```python
avg_over_beta
```

```python
plt.plot(avg_over_beta, label=r'$\mu$' + '={}'.format(MU))
plt.title(r'$\rho$' + ' values for different ' + r'$\beta$', fontsize=15)
plt.xlabel(r'$\beta$', fontsize=15)
plt.ylabel(r'$\rho$', fontsize=15)
plt.legend(loc='lower right')
plt.tight_layout()
plt.style.use('ggplot')
```

### Full computation
We need now to iterate over mu and beta

```python
mu_list = [0.1, 0.9]
beta_list = [0.1, 0.4, 0.9]
values_per_mu = []
```

```python
for mu in mu_list:
    values_per_mu.append(iterate_over_beta(g, len_g, mu, beta_list)) 
```

```python
for i in range(len(mu_list)):
    plt.plot(values_per_mu[i], label=r'$\mu$' + '={}'.format(mu_list[i]))
    plt.title(r'$\rho$' + ' in function of ' + r'$\beta$', fontsize=15)
    plt.xlabel(r'$\beta$', fontsize=15)
    plt.ylabel(r'$\rho$', fontsize=15)
    plt.legend(loc='lower right')
    plt.tight_layout()
    plt.style.use('ggplot')
    
```
